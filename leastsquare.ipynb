solve using least squares<br>
x -1 0 1 2 3<br>
y 0 1 2 9 28<br>
fit a linear and quadratic approximation


```python
#linear approximation
import numpy as np
import matplotlib.pyplot as plt
x=[-1,0,1,2,3]
y=[0,1,2,9,28]
a=[1 for i in range(len(x))]
A1=np.array((a,x))
A=np.transpose(A1)
b1=np.array((y))
b=np.transpose(b1)
#Ax=b has no solution multiply Atranspose on both sides
AT=np.transpose(A)
A=np.matmul(AT,A)
b=np.matmul(AT,b)
z = np.linalg.solve(A, b)
v=np.arange(-1,3.1,.1)
plt.plot(v,z[0]+z[1]*v,label="y=a+bx")
plt.scatter(x,y,color='blue',marker='.')
plt.title("Linear approximation")
plt.legend()
```

```python
#quadratic approximation
import numpy as np
import matplotlib.pyplot as plt
x=[-1,0,1,2,3]
x2=[i*i for i in x]
y=[0,1,2,9,28]
a=[1 for i in range(len(x))]
A1=np.array((a,x,x2))
A=np.transpose(A1)
b1=np.array((y))
b=np.transpose(b1)
#Ax=b has no solution multiply Atranspose on both sides
AT=np.transpose(A)
A=np.matmul(AT,A)
b=np.matmul(AT,b)
z = np.linalg.solve(A, b)
v=np.arange(-1,3.1,.1)
plt.plot(v,z[0]+z[1]*v+z[2]*(v**2),label="y=a+bx+cx**2")
plt.scatter(x,y,color='blue',marker='.')
plt.title("Quadratic approximation")
plt.legend()
```
